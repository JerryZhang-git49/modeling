```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

设置中文格式和样式

```python
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")
```

生成数据

```python
np.random.seed(42)  # 设置随机种子以确保结果可重复
n_samples = 200
X_key = np.random.randn(n_samples, 8)  # 生成关键特征数据
true_coefficients = np.array([2.5, -1.8, 3.2, 1.5, -2.1, 2.8, -1.3, 1.9])
X_noise = np.random.randn(n_samples, 2)  # 生成噪声特征数据
y_true = np.dot(X_key, true_coefficients) + np.random.normal(0,0.5,size=n_samples)  # 生成目标变量数据
X_all = np.concatenate((X_key, X_noise), axis=1)  # 合并关键特征和噪声特征

feature_name = [f'关键变量{i+1}' for i in range(8)] + [f'噪声变量{i+1}' for i in range(2)]
df = pd.DataFrame(X_all, columns=feature_name)
df['目标变量'] = y_true

print("数据集基本信息:")
print(f"样本数量: {n_samples}")
print(f"特征数量: {X_all.shape[1]} (8个关键变量 + 2个误差变量)")
print(f"目标变量统计: 均值={y_true.mean():.2f}, 标准差={y_true.std():.2f}")
print("\n数据集前5行:")
print(df.head())
```

逐步添加关键变量的回归分析

```python
def progressive_regression_analysis(X,y,feature_names,title_prefix=""):
    """
    逐步回归分析函数
    :param X: 特征数据
    :param y: 目标变量
    :param feature_names: 特征名称列表
    :param title_prefix: 图表标题前缀
    """
    results = []
    predictions = []

    for i in range(1,X.shape[1]+1):
        X_subset = X[:,:i]   #读取前i个变量
        
        # 训练线性回归模型
        model = LinearRegression()
        model.fit(X_subset,y)

        # 预测
        y_pred = model.predict(X_subset)
        predictions.append(y_pred)

        # 计算评估指标
        mse = mean_squared_error(y,y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y,y_pred)
        r2 = r2_score(y,y_pred)

        # 保存结果
        results.append({
            '变量数量': i,
            '使用变量': feature_names[:i],
            '均方误差(MSE)': mse,
            '均方根误差(RMSE)': rmse,
            '平均绝对误差(MAE)': mae,
            'R²': r2
        })

    return results, predictions
```

关键变量逐步添加分析

```python
print('\n' + '='*50)
print('关键变量逐步添加分析')
print('='*50)

key_results, key_predictions = progressive_regression_analysis(
    X_key, y_true, [f'关键变量_{i+1}' for i in range(8)], title_prefix="关键变量逐步添加分析"
)

# 显示结果
key_results_df = pd.DataFrame(key_results)
print(key_results_df[['变量数量', '均方误差(MSE)', '均方根误差(RMSE)', '平均绝对误差(MAE)', 'R²']].round(4))
```

添加误差变量的分析

```python
print('\n' + '='*50)
print('关键变量与噪声变量逐步添加分析')
print('='*50)

X_all = np.concatenate((X_key, X_noise), axis=1)
noise_results, noise_predictions = progressive_regression_analysis(
    X_all, y_true, [f'关键变量_{i+1}' for i in range(8)] + [f'噪声变量_{i+1}' for i in range(2)], title_prefix="关键变量与噪声变量逐步添加分析"
)

# 只显示基础模型和添加误差变量后的结果
noise_results_df = pd.DataFrame(noise_results)
comparison_results = noise_results_df[noise_results_df['变量数量'].isin([8, 9, 10])]
print(comparison_results[['变量数量', '均方误差(MSE)', '均方根误差(RMSE)', '平均绝对误差(MAE)', 'R²']].round(4))
```

可视化分析

```python
fig,axes = plt.subplots(2, 2, figsize=(15, 12))  # 创建2*2子图的画布

# 5.1 绘制关键变量逐步添加的评估指标变化图
ax1 = axes[0,0]  # 选中左上角子图
key_metrics = key_results_df[['变量数量', '均方误差(MSE)', 'R²']]
ax1_twin = ax1.twinx()  # 创建双y轴

# 绘制关键变量逐步添加的评估指标变化图
line1 = ax1.plot(key_metrics['变量数量'],key_metrics['均方误差(MSE)'],'bo-',label='mse',linewidth=2,markersize=6)
line2 = ax1.plot(key_metrics['变量数量'],key_metrics['R²'], 'ro-', label='r2', linewidth=2, markersize=6)

ax1.set_xlabel('使用变量数量')
ax1.set_ylabel('均方误差(MSE)', color='b')
ax1_twin.set_ylabel('R²', color='r')
ax1.set_title('关键变量逐步添加的评估指标变化图', fontsize=14)

# 合并图例
lines1,labels1 = ax1.get_legend_handles_labels()
lines2,labels2 = ax1_twin.get_legend_handles_labels()
ax1.legend(lines1+lines2,labels1+labels2,loc='upper right')


# 5.2 预测vs真实值散点图（最终模型）
ax2 = axes[0, 1]
final_pred_key = key_predictions[-1]
ax2.scatter(y_true, final_pred_key, alpha=0.6, color='green')
ax2.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', linewidth=2)
ax2.set_xlabel('真实值')
ax2.set_ylabel('预测值')
ax2.set_title(f'8个关键变量模型预测效果\nR² = {key_results[-1]["R²"]:.4f}')
ax2.grid(True, alpha=0.3)


# 5.3 误差变量影响对比
ax3 = axes[1, 0]
comparison_data = comparison_results[['变量数量', '均方误差(MSE)', 'R²']]
ax3_twin = ax3.twinx()

x_pos = [0, 1, 2]
x_labels = ['8个关键变量', '+误差变量1', '+误差变量2']

bars1 = ax3.bar([x-0.2 for x in x_pos], comparison_data['均方误差(MSE)'], width=0.4, 
                label='均方误差(MSE)', color='lightcoral', alpha=0.7)
bars2 = ax3_twin.bar([x+0.2 for x in x_pos], comparison_data['R²'], width=0.4, 
                     label='R²', color='lightblue', alpha=0.7)

ax3.set_xlabel('模型配置')
ax3.set_ylabel('均方误差(MSE)', color='red')
ax3_twin.set_ylabel('R²', color='blue')
ax3.set_title('误差变量对模型性能的影响')
ax3.set_xticks(x_pos)
ax3.set_xticklabels(x_labels, rotation=15)

# 添加数值标签
for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
    height1 = bar1.get_height()
    height2 = bar2.get_height()
    ax3.text(bar1.get_x() + bar1.get_width()/2., height1 + 0.01,
             f'{height1:.3f}', ha='center', va='bottom', fontsize=8)
    ax3_twin.text(bar2.get_x() + bar2.get_width()/2., height2 + 0.01,
                  f'{height2:.3f}', ha='center', va='bottom', fontsize=8)

lines3, labels3 = ax3.get_legend_handles_labels()
lines4, labels4 = ax3_twin.get_legend_handles_labels()
ax3.legend(lines3 + lines4, labels3 + labels4, loc='center right')


# 5.4 残差分析
ax4 = axes[1, 1]
residuals_key = y_true - final_pred_key
residuals_noise = y_true - noise_predictions[-1]

ax4.scatter(final_pred_key, residuals_key, alpha=0.6, label='仅关键变量', color='green')
ax4.scatter(noise_predictions[-1], residuals_noise, alpha=0.6, label='包含误差变量', color='red')
ax4.axhline(y=0, color='black', linestyle='--', alpha=0.8)
ax4.set_xlabel('预测值')
ax4.set_ylabel('残差')
ax4.set_title('残差分析对比')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
